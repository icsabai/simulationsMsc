```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22])
#y = np.array([100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100])
y = np.random.rand(22)
```

```python
idx0 = np.array([2,15])
idx1 = np.array([2,8,15])
idx2 = np.array([2,7,15])
idx3 = np.array([2,9,15])
idx4 = np.array([2,6,15])

mymodel1 = np.poly1d(np.polyfit(x[idx1], y[idx1], 2))
mymodel2 = np.poly1d(np.polyfit(x[idx2], y[idx2], 2))
mymodel3 = np.poly1d(np.polyfit(x[idx3], y[idx3], 2))
mymodel4 = np.poly1d(np.polyfit(x[idx4], y[idx4], 2))

myline = np.linspace(1, 22, 100)

plt.figure(figsize=(3,3))
plt.scatter(x[idx0], y[idx0],s=100)
plt.plot(myline, mymodel3(myline))
plt.plot(myline, mymodel1(myline))
plt.plot(myline, mymodel2(myline))
plt.plot(myline, mymodel4(myline))
plt.xlim(0,23)
plt.title("$y=a_1+a_2x+a_3x^2$")
plt.ylim(-0.5,0.5)
```
